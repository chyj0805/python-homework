```python
filename = 'budget_data.csv'

# Open the file for reading
with open(filename, 'r') as file:
    # Read all lines from the file
    lines = file.readlines()

data = []  # List to store the dictionaries

# Process each line and convert to dictionary
for line in lines[1:]:  # Skip the first line (header)
    line = line.strip()  # Remove extra newline characters
    values = line.split(',')  # Split the line by comma (assuming CSV format)
    entry = {'Date': values[0], 'Profit/Losses': int(values[1])}  # Create a dictionary with the relevant keys and values
    data.append(entry)  # Add the dictionary to the list

# Display all rows in the data as dictionaries
for entry in data:
    print(entry)

```
